### Задание 16

```python
x = 8.52
c = 9
b = x + np.power(c,2)
a = np.power(np.sqrt(b + c),3)
y = np.power(np.cos(b),2) + b * np.power(np.cos(np.power(a,2)),4)

print(f'y = {y}')
print(f'a = {a}')
print(f'b = {b}')
```

```python
class Calculator:
    def __init__(self, x, c):
        self.x = x
        self.c = c
        self.b = self.calculate_b()
        self.a = self.calculate_a()
        self.y = self.calculate_y()

    def calculate_b(self):
        return self.x + np.power(self.c, 2)

    def calculate_a(self):
        return np.power(np.sqrt(self.b + self.c), 3)

    def calculate_y(self):
        return np.power(np.cos(self.b), 2) + self.b * np.power(np.cos(np.power(self.a, 2)), 4)

    def display_results(self):
        print(f'y = {self.y}')
        print(f'a = {self.a}')
        print(f'b = {self.b}')

    def plot_graph(self):
        c_values = np.linspace(0.1, 5, 100)
        b_values = self.x + np.power(c_values, 2)  # Исправленная строка
        a_values = np.power(np.sqrt(b_values + c_values), 3)
        y_values = np.power(np.cos(b_values), 2) + b_values * np.power(np.cos(np.power(a_values, 2)), 4)

        plt.figure(figsize=(12, 8))

        plt.subplot(3, 1, 1)
        plt.plot(c_values, b_values, label='b = x + c^2', color='blue')
        plt.title('График функции b')
        plt.xlabel('c')
        plt.ylabel('b')
        plt.grid()
        plt.legend()

        plt.subplot(3, 1, 2)
        plt.plot(c_values, a_values, label='a = (√(b + c))^3', color='green')
        plt.title('График функции a')
        plt.xlabel('c')
        plt.ylabel('a')
        plt.grid()
        plt.legend()

        plt.subplot(3, 1, 3)
        plt.plot(c_values, y_values, label='y = cos^2(b) + b * cos^4(a^2)', color='red')
        plt.title('График функции y')
        plt.xlabel('c')
        plt.ylabel('y')
        plt.grid()
        plt.legend()

        plt.tight_layout()
        plt.show()

x = 8.52
c = 9
calculator = Calculator(x, c)
calculator.display_results()
calculator.plot_graph()
```
