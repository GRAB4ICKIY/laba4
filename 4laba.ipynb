```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = 1.3
k = 4

a = np.log10(np.abs(x))
b = np.exp(2 * x) + a * x
y = x * np.power(a,3) + np.power(b,2)

print(f'y = {y}')
print(f'a = {a}')
print(f'b = {b}')
```

```python
class Calculator:
    def __init__(self, x, k):
        self.x = x
        self.k = k
        self.a = self.calculate_a()
        self.b = self.calculate_b()
        self.y = self.calculate_y()

    def calculate_a(self):
        return np.log10(np.abs(self.x))
    
    def calculate_b(self):
        return np.exp(2 * self.x) + self.a * self.x
    
    def calculate_y(self):
        return self.x * np.power(self.a,3) + np.power(self.b,2)

    def display_results(self):
        print(f'y  = {self.y}')
        print(f'a  = {self.a}')
        print(f'b  = {self.b}')

    def plot_graph(self):

        x_values = np.linspace(0.1, 5, 100) 
        a_values = np.log10(np.abs(x_values))
        b_values = np.exp(2 * x_values) + a_values * x_values
        y_values = x_values * np.power(a_values, 3) + np.power(b_values, 2)


        data = pd.DataFrame({
            'x': x_values,
            'a': a_values,
            'b': b_values,
            'y': y_values
        })


        plt.figure(figsize=(12, 8))

 
        plt.subplot(3, 1, 1)
        sns.lineplot(data=data, x='x', y='y', color='blue')
        plt.title('График функции y')
        plt.xlabel('x')
        plt.ylabel('y')
        plt.grid()

        plt.subplot(3, 1, 2)
        sns.lineplot(data=data, x='x', y='a', color='orange')
        plt.title('График функции a')
        plt.xlabel('x')
        plt.ylabel('a')
        plt.grid()

  
        plt.subplot(3, 1, 3)
        sns.lineplot(data=data, x='x', y='b', color='green')
        plt.title('График функции b')
        plt.xlabel('x')
        plt.ylabel('b')
        plt.grid()

        plt.tight_layout()
        plt.show()
```

```python
x = 1.3
k = 4
calculator = Calculator(x, k)
calculator.display_results()
calculator.plot_graph()
```

```python
t = 4.1
p = 3

k = np.sqrt(p * t)
x = p * np.power(t , 2) + np.sqrt(k)
y = np.tan(3) * np.power(x , 2) + k * t 

print(f'k = {k}')
print(f'x = {x}')
print(f'y = {y}')
```

```python
class Calculator:
    def __init__(self, p, t):
        self.t = t
        self.p = p
        self.k = self.calculate_k()
        self.x = self.calculate_x()
        self.y = self.calculate_y()

    def calculate_k(self):
        return np.sqrt(self.p * self.t)
    
    def calculate_x(self):
        return  self.p * np.power(self.t , 2) + np.sqrt(self.k)
    
    def calculate_y(self):
        return np.tan(3) * np.power(self.x , 2) + self.k * self.t 

    def display_results(self):
        print(f'k  = {self.k}')
        print(f'x  = {self.x}')
        print(f'y  = {self.y}')

    def plot_graph(self):

        t_values = np.linspace(0.1, 5, 100) 
        k_values = np.sqrt(self.p * t_values)
        x_values = self.p * np.power(t_values , 2) + np.sqrt(k_values)
        y_values = np.tan(3) * np.power(self.x , 2) + k_values * t_values 


        data = pd.DataFrame({
            't': t_values,
            'k': k_values,
            'x': x_values,
            'y': y_values
        })


        plt.figure(figsize=(12, 8))

 
        plt.subplot(3, 1, 1)
        sns.lineplot(data=data, x='t', y='k', color='blue')
        plt.title('График функции k')
        plt.xlabel('t')
        plt.ylabel('k')
        plt.grid()

        plt.subplot(3, 1, 2)
        sns.lineplot(data=data, x='t', y='x', color='orange')
        plt.title('График функции x')
        plt.xlabel('t')
        plt.ylabel('x')
        plt.grid()

  
        plt.subplot(3, 1, 3)
        sns.lineplot(data=data, x='t', y='y', color='green')
        plt.title('График функции y')
        plt.xlabel('t')
        plt.ylabel('y')
        plt.grid()

        plt.tight_layout()
        plt.show()
```
